```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import graphviz
from sklearn.model_selection import train_test_split
```

```python
from sklearn import datasets
iris = datasets.load_iris()

df_X = pandas.DataFrame(iris["data"], columns = iris["feature_names"])
df_X.head()
```

```python
df_Y = pandas.DataFrame(iris["target"], columns = ["target_names"])
df_Y.head()
```

```python
print(iris["target_names"])
```

```python
df = pandas.concat([df_X,df_Y],axis = 1)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, test_size=0.3, random_state=1)
```

```python
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (20, 20)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
data = tree.export_graphviz(dtree, out_file=None, feature_names=iris["feature_names"])
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img = pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
from sklearn import metrics
y_pred = dtree.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
